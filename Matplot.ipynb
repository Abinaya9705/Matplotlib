```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()
```

```python
fig3=plt.figure
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,2,2])
languages=['English','French','Spanish','Latin','German']
people=[100,200,300,400,500]
myexplode=[0.2,0.2,0,0,0]
color=['aquamarine','lavenderblush','magenta','pink','cyan','alpha']
ax.pie(people,labels=languages,colors=color,autopct='%1.2f%%',explode=myexplode)
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(-5, 5, 50)
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add gridlines
plt.grid(True)

# Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Gridlines')

# Show the plot
plt.show()
```
